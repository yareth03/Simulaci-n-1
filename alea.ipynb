```python
from tabulate import tabulate

def randomyare():
    xi = int(input("Escribe el valor inicial: "))
    a = int(input("Escribe el multiplicador constante: "))
    c = int(input("Escribe el incremento: "))
    m = int(input("Escribe el modulo: "))
    n = int(input("¿Cuántas iteraciones quieres? "))
    tabla = [ ]
    p = 0


    for i in range(1, n + 1):

        xi = (a * xi + c) % m
        p = float(xi)
        tabla.append([i, xi, xi/m])



    return tabla,n


tb, n = randomyare()
print(tabulate(tb, tablefmt="grid"))

```
